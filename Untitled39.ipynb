```python
from keras.applications import vgg16
```

```python
vgg16_model = vgg16.VGG16()
```

```python
vgg16_model.summary()
```

```python
n = vgg16_model.layers.pop()
vgg16_model.summary()
```

```python
new_model = Sequential()
for layer in vgg16_model.layers[:-1]:
    new_model.add(layer)
    
new_model.summary()
```

```python
for layer in new_model.layers:
    layer.trainable = False
```

```python
new_model.add(Dense(9, activation='softmax'))
```

```python
new_model.summary()
```

```python
new_model.compile(Adam(learning_rate=.0001,),loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
train_path = 'hcaptcha_labels/data/train/'
valid_path = 'hcaptcha_labels/data/valid/'
```

```python
training_batch = ImageDataGenerator().flow_from_directory(train_path,target_size=(224,224),classes=list(labels),batch_size=32)
vlidtion_batch = ImageDataGenerator().flow_from_directory(valid_path,target_size=(224,224),classes=list(labels),batch_size=4)
```

```python
new_model.fit_generator(training_batch, validation_data=vlidtion_batch,)
```

```python
for label in labels:
    num_total_data = len(os.listdir(f'hcaptcha_labels/{label}/'))
    num_training_data = int(np.round(.75*num_total_data))
    num_valid_data = num_total_data-num_training_data
    counter = 0
    for img in os.listdir(f'hcaptcha_labels/{label}/'):
        if counter < num_training_data:
            shutil.copy(f'hcaptcha_labels/{label}/{img}',f'hcaptcha_labels/data/train/{label}/')
        else:
            shutil.copy(f'hcaptcha_labels/{label}/{img}',f'hcaptcha_labels/data/valid/{label}/')
        counter += 1
```
